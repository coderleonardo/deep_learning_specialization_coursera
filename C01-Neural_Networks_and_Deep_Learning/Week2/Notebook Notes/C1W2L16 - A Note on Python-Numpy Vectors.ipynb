```python
import numpy as np
```

```python
a = np.random.randn(5)
```

```python
print(a)
```

```python
print(a.shape)
```

```python
print(a.T)
```

```python
print(np.dot(a, a.T))
```

```python
# Vemos dos resultados acima que nosso elemento "a" nao se comporta como um vetor linha,
# que era o que esperavamos. Portanto, vamos evitar usar coisas desse tipo
# Vamos usar de fato a seguinte implementacao, que eh mais intuitiva e retorna o esperado

a = np.random.randn(5,1)
```

```python
print(a)
```

```python
print(a.T)
```

```python
print(np.dot(a, a.T))
```

```python
# Os resultados acima sao como esperavamos!

assert(a.shape == (5, 1))
# Mais sobre o comando "assert" em Python aqui: https://www.educative.io/edpresso/what-does-assert-mean-in-python?
```
