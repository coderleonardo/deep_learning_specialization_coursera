```python
import numpy as np
import random 
```

```python
A = np.array([[56.0, 0.0, 4.4, 68.0],
              [1.2, 104.0, 52.0, 8.0],
              [1.8, 135.0, 99.0, 0.9]])

print(A)
print(A.shape)
```

```python
cal = A.sum(axis=0) # Soma as linhas (axis=0) e retorna o vetor linha referente
print(cal)
```

```python
# Broadcasting
percentage = A / cal.reshape(1,4)
```

```python
print(percentage)
```

```python
'''
Broadcasting em Python Ã© tomar uma matriz ou vetor e fazer operacoes entre ela e um vetor de tamanho diferente.
Exemplos:
1. [[1],
    [2],
    [3]] + 100
    =
    [[1],
    [2],
    [3]] 
    + 
    [[100],
    [100],
    [100]]

2. [[1, 2, 3],
    [4, 5, 6]] + [[100, 200, 300]]
    =
    [[1, 2, 3],
    [4, 5, 6]]
     + 
    [[100, 200, 300],
    [100, 200, 300]]

Python Numpy altera o shape (reshape) da matriz/vetor para calcular as operacoes elemento a elemento
'''
```

```python
# Exemplo 1
a = np.random.randn(2, 3)
display(a)
print(a.shape)
b = np.random.randn(2, 1)
display(b)
print(b.shape)
c = a + b
display(c)
print(c.shape)
```

```python
# Exemplo 2
try:
  a = np.random.randn(4, 3)
  display(a)
  print(a.shape)
  b = np.random.randn(3, 2)
  display(b)
  print(b.shape)
  c = a + b
  display(c)
  print(c.shape)
except ValueError:
  print("ValueError: operands could not be broadcast together with shapes")
```
